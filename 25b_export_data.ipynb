# Lesson 25b: Export data

```python
import pandas as pd
import numpy as np
```

```python
df_blood = pd.DataFrame({"Group" : ["0","0","A","A","B","B","AB","AB"],
                        "Rh"     : ["+","-","+","-","+","-","+","-"],
                        "Population": [31,6,35,15,8,55,3,14]})
```

```python
df_blood
```

```python
df_blood.to_csv("blood.csv")
```

```python
pd.read_csv("blood.csv")
# Note that there is something wrong with index
```

```python
# If we want to divide all values in one column by 100:

df_blood.loc[:, "Population"] /= 100

df_blood
```

```python
# Let us write it down again and set that we do not want to export index:
df_blood.to_csv("blood.csv", index = False)
```

```python
df_blood.to_csv("blood.csv", index = False, columns = ["Group","Rh"], sep = ";")
```

```python
# Note that when ";" was chosen before as separator, now I also have to do it.
pd.read_csv("blood.csv", sep = ";")
```

```python
# To append the data to the existing file we use the property "mode=a"
df_blood.to_csv("blood.csv", index = False, columns = ["Group","Rh"], sep = ";", mode = "a", header = False)
```

```python
pd.read_csv("blood.csv", sep = ";")
# so we doubled the data.
```
