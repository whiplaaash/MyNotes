# Lesson 20a: Data Modification

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as math
```

## Load data

```python
frame = pd.read_csv("Insurance.csv")
frame.head(6)
```

## Modification of the cell

```python
# We will be modifying data, so we need to point a row and a column (or just a cell) where the modification 
# will be done:

frame.loc[2, "Claims"]
```

```python
# I want to introduce there a value which is smaller by 1 from the original one:

frame.loc[2, "Claims"] = 19
frame.head()
```

```python
# Now we increase the value by one:

frame.loc[2, "Claims"] += 1
frame.head()
```

```python
# To check a condition within a column we use

isYounger25 = frame["Age"] == "<25"
frame[isYounger25]
```

```python
# or

frame.where(isYounger25).dropna()

# If dropna() is not added, then full frame will be shown with all irrelevant data.
```

```python
# We will be now modifying the column "Holders", BUT to find the ones that we are interested in
# we will use a kind of index, which is "isYounger25" and not the number of a row.

frame.loc[isYounger25, "Holders"]
```

```python
# And now we add 100 to all these rows:

frame.loc[isYounger25, "Holders"] + 100
```

```python
# Now we add the modified column to our data frame:

frame.loc[isYounger25, "Holders"] = frame.loc[isYounger25, "Holders"] + 100
frame.head()
```

```python
# It can be done in a simpler way:

frame.loc[isYounger25, "Holders"] = frame["Holders"] + 1000
frame.head()
```
