## Adding and removing columns

```python
frame_1 = pd.read_csv("Mcdonalds.csv", usecols = ["Item", "Serving Size", "Calories", "Saturated Fat", 
                                                  "Cholesterol", "Sugars"], index_col = "Item")
```

```python
frame_1["Saturated Fat"].head()
```

```python
# To add a column I need:

frame_1["New Column"] = 0
```

```python
frame_1.head()
```

```python
# Adding values of two columns and showing them in a new column:

CholesterolSugars = frame_1.Cholesterol + frame_1.Sugars
frame_1["CholesterolSugars"] = CholesterolSugars

frame_1
```

```python
# If I want to add a column not at the end of the frame but in the middle:

frame_1.insert(loc = 2, column = "CSS", value = frame_1.Calories + frame_1.Sugars)
frame_1
```

```python
# Removing a column:

del frame_1["New Column"]
```

```python
frame_1.head()
```

```python
# Another way:

frame_1 = frame_1.drop(columns = "CholesterolSugars")
```

```python
frame_1
```

```python
# I can add inplace=True to make this operation inside the frame.

frame_1.drop(columns = "CSS", inplace=True)
```

## Cleaning the data

### How to remove the data: using "dropna()"

```python
frame.head()
```

```python
# First we change some values to NaN to remove them later:

frame.loc[1, "Calories"] = np.NaN
frame.loc[2, "TotalFat"] = np.NaN

frame.head()
```

```python
frame.dropna().head()  # By using this I remowve the whole rows which contain at least one NaN.
```

```python
frame.dropna(how = "all").head()  
# By changing the parameter "how" I can drop only these rows which have NaN in all places.
```

```python
frame.dropna(subset = "Calories").head()
# By changing subset, I can choose which set of columns this function should look at.
```

```python
frame.dropna(axis = 1).head()  # I can also choose that whole columns with NaN can be reomoved.
```

### How to add the data: using "fillna()"

```python
frame.head()
```

```python
# First, I spoil the frame a bit:

frame.loc[1, "Category"] = np.NaN
frame.loc[4, "Calories"] = np.NaN
frame.loc[1, "Saturated Fat"] = np.NaN
```

```python
frame.head()
```

```python
frame.fillna(value = 0).head()   # I replace NaN by any value, in this case by 0, in whole frame.
```

```python
# We see this is not a good solution because, for example, Category cannot have the value 0.
# To deal with it, we declare a dictionary and define what insert where.

replace_dic = {"Category": "UNKNOWN", "Calories": 0, "Saturated Fat": 0, "TotalFat": 10}
```

```python
frame.fillna(value = replace_dic).head()
```

```python
# We can also work on a chosen one column:

frame["Category"].fillna(value = "Something", inplace = True)

frame.head()
```

```python
frame["Calories"].fillna(value = 0, inplace = True)
frame.head()
```

```python
# I can replace NaN by the mean of the column:

avg_total_fat = frame["TotalFat"].mean()
```

```python
frame["TotalFat"].fillna(value = avg_total_fat, inplace = True)
frame.head()

frame["TotalFat"] = round(frame["TotalFat"],2)

frame.head()
```

```python
# I can also replace NaN by copying the value of the previous row. 
# For this I can use the parameter "method=ffill":

frame["Saturated Fat"].fillna(method = "ffill", inplace = True)
frame.head()
```
