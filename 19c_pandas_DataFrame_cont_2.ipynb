## Data filtering

### Using conditions with some operators

```python
frame = pd.read_csv("Mcdonalds.csv", usecols = ["Item", "Category", "Serving Size", "Calories", "TotalFat"])
frame.head()
```

```python
# Let us show all values that have calories bigger than 400:

(frame["Calories"] >= 400).head()
```

```python
# Better way by declaring a variable:

Calories_big = frame["Calories"] >= 400

# Now only those values that satisfy the condition are shown:

frame[Calories_big].head()
(frame["Calories"] >= 400).head()
```

```python
# Another example

is_Breakfast = frame['Category'] == "Breakfast"

frame[is_Breakfast].head(8)
```

```python
# Showing values if two conditions are satisfied:

frame[Calories_big & is_Breakfast].head(8)
```

```python
# Showing an opposite condition (the negation operator):

frame[~ is_Breakfast].head()
```

```python
# Different example:

color = ["blue", "blue", "red", "red", "white", "brown", "brown"]
size = ["S", "M", "M", "L", "S", "L", "S"]
gender = ['w', 'w', 'm', 'm', 'u', 'w', 'm']

clothes = pd.DataFrame({"color": color, "size": size, "gender": gender})
clothes
```

```python
is_color_blue = clothes['color'] == "blue"
is_color_red = clothes['color'] == "red"
is_color_brown = clothes['color'] == "brown"
is_size_s = clothes['size'] == "S"
is_size_m = clothes['size'] == "M"
is_gender_w = clothes['gender'] == "w"
is_gender_m = clothes['gender'] == "m"
```

```python
clothes[is_color_blue & is_size_m & is_gender_w]
```

```python
clothes[is_color_blue | is_color_brown | is_color_red]
```

```python
# In the above examples the order of condtions was not important.
# Below "&" is always performed as first and "|" later.

clothes[is_color_blue & is_size_m | is_gender_w]
```

```python
# Note that "&" = multiplication (is first), and "|" = addition (is done later)
```

### Using methods where and query

```python
# We already defined the condition before:

Calories_big
```

```python
# But now we use the method "where()"

frame.where(Calories_big).head()

# Note that here the rows where the condition is not satisfied display NaN, 
# in the previous method these rows were not shown at all.

# I can get the same shape of the frame if I attach ".dropna(how = "all")" to the last case.
```

```python
# The method "query()" is exactly as it works in SQL, and to use it we need to write
# an expression in it:

frame.query("Category == 'Breakfast'").head(10)
```

```python
# Next example:

frame.query("Category in ['Desserts', 'Beverages']").value_counts()
```

```python
# Next example:

frame.query("Category == 'Desserts' and Calories <= 200")
```

## Methods: isin, isnull, notnull, between

```python
# "isin()" can be used instead of the logical operator "or" (|):

# Let us show it for the frame "clothes":

clothes
```

```python
# Here we check the color:

col = clothes["color"].isin(["blue","red", "brown"])
col
```

```python
# Here we show all rows for these colors:

clothes[col]
```

```python
# Now I spoil my table to show how "isnull()" works:

clothes.loc[6, "size"] = np.NaN
clothes
```

```python
# Now I can check which rows contain NaN:

has_size = clothes["size"].notnull()

has_size
```

```python
# and next I can show only these values that have defined size:

clothes[has_size]
```

```python
# "isnull()" works oppositely to "notnull()" and returns only these values where the size is not given:

no_size = clothes["size"].isnull()
no_size
```

```python
# Now I can show these values:

clothes[no_size]
```

```python
# To show how "between" works we get back to the frame mcdonald.

# We want to show products that contain more than 300 kal and less than 400 cal:

less_400 = frame["Calories"] <= 400
more_300 = frame["Calories"] >= 300

frame[less_400 & more_300].head()
```

```python
# Now we use "between()":

between_cal = frame["Calories"].between(300,400)
frame[between_cal].head()
```

## Duplicates

```python
# Checking if values are unique:

clothes
```

```python
clothes["size"].fillna(value = "M", inplace = True)
clothes
```

```python
clothes["color"].is_unique
```

```python
# Showing unique values:

clothes["color"].unique()
```

```python
# Counting unique values:
clothes["color"].nunique()
```

```python
# or:
len(clothes["color"].unique())
```

```python
# What if I have NaN somewhere?

clothes.loc[6, "color"] = np.NaN
clothes
```

```python
clothes["color"].unique()
# Note that NaN is here
```

```python
# so by using len() we will count this row:

len(clothes["color"].unique())
```

```python
# With "nunique()" this problem does not exists:
clothes["color"].nunique()
```

```python
# Another way to show not unique values:

not_unique_values = clothes["color"].duplicated(keep = "first")

clothes[not_unique_values]
```

```python
# To show unique values:

clothes[~ not_unique_values]
```

```python
# To remove duplicated values:

clothes.drop_duplicates(subset = "color", keep = "first")
```

```python
# If the subset is bigger than one column, then the metod removes the row for which 
# both the color AND the size are duplicated:

clothes.drop_duplicates(["color", "size"])
```
