## Import libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

# There is an error in "statsmodels" library as some function was removed. To fix it and be able
# to display a summary of data we need to use:
from scipy import stats 
stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq.df)
```

## Upload data

```python
raw_data = pd.read_csv("2.02.Binary_predictors.csv")
raw_data
```

```python
# Now we convert cathegorical variables into numerical ones.

data = raw_data.copy()
data["Admitted"] = data["Admitted"].map({"Yes":1, "No":0})
data["Gender"] = data["Gender"].map({"Female":1, "Male":0})
data
```

## Variables

```python
y = data["Admitted"]
x1 = data["Gender"]
x2 = data[["SAT", "Gender"]]
```

## Regression

```python
x = sm.add_constant(x1)

reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
results_log.summary()
```

```python
# To estimate the significance of the model we need to calculate: Log(odds2/odds1) = b1*(Gender2-Gender1),
# where b1 is taken from the table and b1=2.08. This gives: odds2 = 7.99 odds1. It means that females have
# 8 times higher chances to get admitted than men.
```

```python
# We can repeat the above analysis including "SAT" to x1 and compare the results.

w = sm.add_constant(x2)

reg_log = sm.Logit(y,w)
results_log = reg_log.fit()
results_log.summary()
```

## Accuracy

```python
# Values predicted by the model:

np.set_printoptions(formatter = {"float": lambda w: "{0:0.2f}".format(w)})
results_log.predict()
```

```python
# Actual values:

np.array(data["Admitted"])
```

```python
# To chck the accuracy we need to compare the predicted and actual values in a efficient way.
# Here it is:

results_log.pred_table()

# To make it more readable we modify the table:

cm_df = pd.DataFrame(results_log.pred_table())
cm_df.columns = ["Predicted 0", "Predicted 1"]
cm_df = cm_df.rename({0:"Actual 0", 1:"Actual 1"})
cm_df

# The matrix we get is called the confusion matrix.
```

```python
# It means that our model predicted correctly 69 entries with 0 and 90 entries with 1. It gave 5 entries with 1
# while it should be 0 and 4 entries with 0 while it should be 1. In total, it predicted 159 entries correctly,
# and 9 entries incorrectly. 

# The accuracy is calculated as 159/(159+9)=0.946. It is very good.

# There is a simple code to calculate the accuracy:

cm = np.array(cm_df)
accuracy_train = (cm[0,0] + cm[1,1])/cm.sum()
accuracy_train
```

```python
test_actual = test["Admitted"]
test_data = test.drop("Admitted", axis = 1)  # This object includes everything apart from "Admitted".
test_data = sm.add_constant(test_data)
# test_data = test_data[w.columns.values]    # If I need to change the order of columns.
test_data
```

```python
# We need to creat the confusion matrix, but unfortunately statsmodels do not provide a simple method.
# We define it as follows:

def confusion_matrix(data, actual_values, model):
    pred_values = model.predict(data)
    bins = np.array([0, 0.5, 1])
    cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
    accuracy = (cm[0,0] + cm[1,1])/cm.sum()
    return cm, accuracy
```

```python
cm = confusion_matrix(test_data, test_actual, results_log)
cm
```
