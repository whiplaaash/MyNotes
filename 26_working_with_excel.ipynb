# Lesson 26: Working with Excel

```python
import pandas as pd
import numpy as np
```

```python
# I can easily read an excel file. But because I only need to work with one sheet, I have to specify its name:
# "sheet_name="name_of_sheet"" (or I can give the number of sheet, here =1)

pd.read_excel("PC_Vehicles-in-use.xlsx", sheet_name = 1)
```

```python
# Or I can work with 3 sheets, and then I have to specify a list, but the outcome will show up as a dictionary:

sheets = pd.read_excel("PC_Vehicles-in-use.xlsx", sheet_name = [1,2,3])
sheets
```

```python
# To call an element of the dictionary:

sheets[1]

# Note that "1" is a key here, and I used it because it was also use when importing the file.
# If I use names of the sheets when importing data, I should also use these names as keys.
```

```python
# Exporting excel data is not as easy as csv files. We need to first create an object which will be
# writing our modified data to new excel file:

excelWriter = pd.ExcelWriter("cars.xlsx")

# Now I can export data:

sheets[1].to_excel(excelWriter, index = False, columns = ["REGIONS/COUNTRIES",2010], sheet_name = "OldEurope2010")

sheets[1].to_excel(excelWriter, index = False, columns = ["REGIONS/COUNTRIES",2011], sheet_name = "OldEurope2011")

# and necessarily save them:

excelWriter.save()
```
