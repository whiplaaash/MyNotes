# Lesson 20b: Adding and removing rows and columns

## Import libraries and load data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as math

frame = pd.read_csv("sleep_time.csv", index_col = "ID")
frame.head()
```

## Removing columns and rows

```python
# Simple way is to use "del" to remove a column:

del frame["bodywt"]
frame.head()
```

```python
# Both columns and rows can be romved using the function "drop()":

frame.drop(axis = 1, columns=["awake", "brainwt"], inplace = True)
frame.head()
```

```python
# or:

frame.drop(axis = "columns", columns = "sleep_cycle", inplace = True)
frame.head()
```

```python
# or the simplest way is:

frame.drop(columns = "sleep_rem", inplace = True)
frame.head()
```

```python
# for removing rows we use:

frame.drop(axis = 0, labels = [1,2], inplace = True)
frame.head()
```

```python
# or:

frame.drop(axis = "rows", labels = [3,4], inplace = True)
frame.head()
```

```python
# or the simplest way is:

frame.drop(labels = 5, inplace = True)
frame.head()
```

```python
# or even: 

frame.drop(6, inplace = True)
frame.head()
```
