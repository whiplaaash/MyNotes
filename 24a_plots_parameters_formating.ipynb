## Import libraries and load data

```python
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline  # This is added to see plots in Jupyter.

bar = pd.read_csv("weather_barcelona.csv", index_col = "Date")
```

```python
rom = pd.read_csv("weather_rome.csv", index_col = "Date")
```

```python
ams = pd.read_csv("weather_amsterdam.csv", index_col = "Date")
```

```python
bar.head()
```

```python
rom.head()
```

```python
ams.head()
```

```python
# To make an arbitrary plot of some data:

bar.plot()
```

```python
# To plot one series:
bar["TempMax"].plot()
```

```python
# or:
bar.plot(y="TempMax")
```

```python
# For 2 series:

bar[["TempMax","TempMin"]].plot()
```

```python
# or:

bar.plot(y=["TempMax","TempMin"])
```

## Parameters of the plot

```python
# Let us first create a dataframe which contains max temperature of each considered city:

temp_max = pd.DataFrame(index = bar.index)

temp_max["bar"] = bar["TempMax"]
temp_max["rom"] = rom["TempMax"]
temp_max["ams"] = ams["TempMax"]

temp_max.head()
```

```python
temp_max.plot()
```

```python
# To enlarge the figure:

temp_max.plot(figsize = (15,5))
```

```python
# To subgroup figure:
temp_max.plot(figsize = (15,5), subplots = True)
```

```python
# To hide the name of x-axis:
# To subgroup figure:
temp_max.plot(figsize = (15,5), use_index = False)
```

```python
# To add the title of the plot:
temp_max.plot(figsize = (15,5), title = "Maximal temperature by city")
```

```python
# To show/hide the legend:
temp_max.plot(figsize = (15,5), legend = False)
```

```python
# To use logarythmic scale:
temp_max.plot(figsize = (15,5), logy = True)
```

```python
# To faocus on a range of time, here first 2 weeks, and show ticks on x-axis, which are at angle 45:
temp_max[:14].plot(figsize=(15,5), xticks = range(14), rot = 45)
```

```python
# To show on the x-axis a table with corresponding values:
temp_max[:7].plot(figsize=(15,5), table = True)
```
