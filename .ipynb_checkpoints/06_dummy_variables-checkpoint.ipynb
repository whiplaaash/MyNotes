# Lesson 6: Dummy variables in multiple linear regression

## Import packages

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv("1.03.Dummies.csv")
```

```python
raw_data
```

```python
# In the table the third column is the cathegorical variable, and to deal with it in some way, 
# we need to map it to the numerical one.

data = raw_data.copy()
data["Attendance"] = data["Attendance"].map({"Yes": 1, "No": 0})
```

```python
data
```

```python
data.describe()
```

## Regression

```python
y = data["GPA"]
x1 = data[["SAT","Attendance"]]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data["SAT"],y,c=data["Attendance"],cmap="PiYG_r")
yhat_no = 0.6439 + 0.0014*data["SAT"] # The regression line for students who didn't attend more than 75% of classes.
yhat_yes = 0.8665 + 0.0014*data["SAT"] # The regression line for students who attended more than 75% of classes.
plt.plot(data["SAT"],yhat_no,lw=4,c="green",label="regression line: no")  # Plotting the line for "no".
plt.plot(data["SAT"],yhat_yes,lw=4,c="red",label="regression line: yes")  # Plotting the line for "yes".
plt.xlabel("SAT",fontsize=20)
plt.ylabel("GPA",fontsize=20)
plt.show()
```

```python
new_data = pd.DataFrame({"const":1,"SAT":[1700,1670],"Attendance":[0,1]})
# Above I assigned new values to entries in a new data framework.

new_data = new_data[["const","SAT","Attendance"]] # Here I produce a table with my new entries, 
# with 3 columns and 2 rows, according to definitions in my framework.

new_data
```

```python
new_data.rename({0:"Bob",1:"Alice"})  # I rename the names of rows.
```

```python
predictions = results.predict(new_data) # I use the object "results" defined before, 
# to make predictions on GPD.
```

```python
predictions
```

```python
predictionsdf = pd.DataFrame({"Predictions":predictions}) # I include "predictions" to my data framework.

joined = new_data.join(predictionsdf) # I add the newly created column to my short table as the fourth column.

joined.rename({0:"Bob",1:"Alice"}) # I rename the names of rows in my new object "joined".
```
